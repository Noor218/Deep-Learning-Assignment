```python
#import all required libraries
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras import models,layers
from google.colab import files
```

```python
#reading csv that is placed on my gdrive
df=pd.read_csv('/content/drive/MyDrive/Kaggle Dataset/creditcard.csv')
```

```python
df.head()
```

```python
df['Class']=df['Class'].astype('float64')#converting the only int type column to float
```

```python
pos_class=df[df['Class']==1]#selecting all the rows that have Class equals to 1 from Dataset 
```

```python
neg_class=df.sample(pos_class.shape[0])#randomly selecting the Class equals to 0 from Dataset
```

```python
neg_class.shape,pos_class.shape
```

```python
#shuffling the Dataset on basic of even or odd
final=[]
pos=0
neg=0
for i in range(len(pos_class)+len(neg_class)):
  if i%2==0:
    final.append(pos_class.iloc[pos])
    pos+=1
  else:
    final.append(neg_class.iloc[neg])
    neg+=1
```

```python
final=pd.DataFrame(final)
final.head()
```

```python
test_val=int((30/100)*len(final))
train_val=int((50/100)*len(final))
val_val=int((20/100)*len(final))
```

```python
#separating the test train and validation data
train_data=final.iloc[:train_val , :-1]
train_labels=final.iloc[:train_val,-1]


val_data=final.iloc[train_val :train_val+val_val , :-1]
val_labels=final.iloc[train_val :train_val+val_val,-1]


test_data=final.iloc[train_val+val_val: , :-1]
test_labels=final.iloc[train_val+val_val: ,-1]
```

```python
mean=train_data.mean(axis=0)
std=train_data.std(axis=0)

train_data-=mean
val_data-=mean
test_data-=mean


train_data/=std
val_data/=std
test_data/=std
```

```python
model= models.Sequential()
model.add(layers.Dense(16, activation="relu", input_shape=(30,)))
model.add(layers.Dense(8, activation="relu"))
model.add(layers.Dense(6,activation="relu"))

model.add(layers.Dense(1,activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(train_data,train_labels,validation_data=(val_data,val_labels),batch_size=4,epochs=90)
```

```python
model.evaluate(test_data,test_labels)
```
